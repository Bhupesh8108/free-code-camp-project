```python
import numpy as np
def calculate(lis):
    if len(lis) == 9:
        dat = np.array(lis)
        data = dat.reshape(3,3)
        maxi =[list(data.max(axis=0)),list(data.max(axis=1)),data.max()]
        meani =[list(data.mean(axis=0)),list(data.mean(axis=1)),data.mean()]
        stdi =[list(data.std(axis=0)),list(data.std(axis=1)),data.std()]
        vari =[list(data.var(axis=0)),list(data.var(axis=1)),data.var()]
        mini =[list(data.min(axis=0)),list(data.min(axis=1)),data.min()]
        sumi =[list(data.sum(axis=0)),list(data.sum(axis=1)),data.sum()]
        dicti = {'mean': meani,'variance':vari,'standard deviation': stdi,'max': maxi,'min':mini,'sum': sumi}
    else :
        raise ValueError('List must contain nine numbers.')
    return dicti
    
calculate(np.arange(0,9))
```
